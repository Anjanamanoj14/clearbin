```python
import os

# Create our data folders
folders = [
    "data/train/plastic",
    "data/train/glass", 
    "data/train/metal",
    "data/train/paper",
    "data/train/cardboard",
    "data/train/trash",
    "data/val/plastic",
    "data/val/glass",
    "data/val/metal", 
    "data/val/paper",
    "data/val/cardboard",
    "data/val/trash",
]

for folder in folders:
    os.makedirs(folder, exist_ok=True)

print("All folders created!")
print("Structure:")
for folder in folders:
    print(" ", folder)
```

```python
import os

dataset_path = "data/dataset-resized"

classes = os.listdir(dataset_path)
print("Classes found:", classes)
print()

total = 0
for cls in classes:
    count = len(os.listdir(os.path.join(dataset_path, cls)))
    total += count
    print(f"  {cls}: {count} images")

print()
print("Total images:", total)
```

```python
import os

dataset_path = "data/dataset-resized"

# Skip hidden files like .DS_Store
classes = [f for f in os.listdir(dataset_path) 
           if os.path.isdir(os.path.join(dataset_path, f))]

print("Classes found:", classes)
print()

total = 0
for cls in classes:
    count = len(os.listdir(os.path.join(dataset_path, cls)))
    total += count
    print(f"  {cls}: {count} images")

print()
print("Total images:", total)
```

```python
import matplotlib.pyplot as plt
from PIL import Image
import os

dataset_path = "data/dataset-resized"
classes = [f for f in os.listdir(dataset_path) 
           if os.path.isdir(os.path.join(dataset_path, f))]

fig, axes = plt.subplots(2, 3, figsize=(10, 7))
axes = axes.flatten()

for i, cls in enumerate(classes):
    # Get first image from each class
    class_path = os.path.join(dataset_path, cls)
    img_name = os.listdir(class_path)[0]
    img_path = os.path.join(class_path, img_name)
    
    img = Image.open(img_path)
    axes[i].imshow(img)
    axes[i].set_title(cls.upper(), fontsize=13, fontweight='bold')
    axes[i].axis("off")

plt.suptitle("ClearBin — All 6 waste categories", fontsize=15, fontweight='bold')
plt.tight_layout()
plt.show()
```
